```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

# Load & Show Image

```python
image = plt.imread('smarties.png') #loading image with pyplot
plt.imshow(image) #show image, y-axis starts at top-left 
plt.show() #must pass to show result
```

# OpenCV Images in PyPlot (Matplotlib)

```python
#OpenCV images must be converted from BGR to RGB to use in PyPlot
CVimage = cv.imread('smarties.png')
plt.imshow(CVimage) #this will show unexpected result (due to BGR & RGB difference)
plt.show()

BGR_to_RGB = cv.cvtColor(CVimage, cv.COLOR_BGR2RGB) #converting BGR to RGB
plt.imshow(BGR_to_RGB)
plt.show()
```

# Multiple Graphs/Images in single Figure

```python
#creating sample images/graphs
img = np.zeros((500, 500, 3), np.uint8)

img1 = img.copy()
img1[:, 251:] = 255

img2 = img.copy()
img2[:250] = 255

img3 = img.copy()
img3[:100, 201:300] = 255

img4 = img.copy()
img4[201:300, 401:] = 255

img5 = img.copy()
img5[401:, 201:300] = 255

img6 = img.copy()
img6[201:300, :100] = 255
```

```python
#showing images/graphs
plt.subplot(2, 3, 1) #'plt.subplot' to show multiple graphs, arg1=rows, arg2=columns, arg3=position of next graph (plt.imshow()). 
plt.imshow(img1)

plt.subplot(2, 3, 2)
plt.imshow(img2)
plt.xticks([])
plt.yticks([])

plt.subplot(2, 3, 3)
plt.imshow(img3)
plt.xticks([0, 100, 200, 300, 400, 500])
plt.yticks([0, 100, 200, 300, 400, 500])
plt.title('Graph_3')

plt.subplot(2, 3, 4)
plt.imshow(img4)
plt.xticks([0, 100, 200, 300, 400, 500], ['a', 'b', 'c', 'd', 'e', 'f'])
plt.yticks([0, 100, 200, 300, 400, 500])
plt.title('Graph_4')

plt.subplot(2, 3, 5)
plt.imshow(img5)
plt.xticks([0, 100, 200, 300, 400, 500], ['a', 'b', 'c', 'd', 'e', 'f'])
plt.yticks([0, 100, 200, 300, 400, 500], ['x', 'y', 'z', 'x', 'y', 'z'])
plt.title('Graph_5')

plt.subplot(2, 3, 6)
plt.imshow(img6)
plt.xticks([0, 100, 200, 300, 400, 500], ['a', 'b', 'c', 'd', 'e', 'f'])
plt.yticks([0, 100, 200, 300, 400, 500], ['x', 'y', 'z', 'x', 'y', 'z'])
plt.title('Graph_6')
          
plt.show()
```
